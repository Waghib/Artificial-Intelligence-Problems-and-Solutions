```python
#load the data set
data = pd.read_csv('House_Selling.csv')
print(data.head()) #display first few rows
print(data.info())
```

```python
X= data[['price', 'Size of house (sqft)']]
```

```python
#using elbow method
wcss = []
for i in range(1, 11):
    k_means = KMeans(n_clusters=i, n_init=10)
    k_means.fit(X)
    wcss.append(k_means.inertia_)

# the elbow curve
plt.plot(range(1, 11), wcss, linewidth=2.5, color="red", marker="8")
plt.title("Elbow Curve")
plt.xlabel("Number of clusters (values of k)")
plt.ylabel("WCSS values")
plt.xticks(np.arange(1, 11, 1))
plt.show()
```

```python
# taking k = 5
k_means = KMeans(n_clusters=5)
# fitting the input data
k_means.fit(X)
# predicting the labels of the input data
y_pred = k_means.predict(X)
# adding the labels to a column named label
df["label"] = y_pred
df.head()
```

```python
# scatterplot of the clusters using seaborn
colors = ["red", "green", "blue", "gold", "purple"]
sns.scatterplot(x="price", y="Size of house (sqft)", hue="label", palette=colors, legend="full", data=df, s=10)
plt.xlabel("price"), plt.ylabel("Size of house (sqft)")
plt.title("Price vs Size of house (sqft)")
plt.show()
```

```python
# adding the new column label into csv file
df.to_csv('House_Selling.csv', index=False)
```

# TASK: 1 (Part 1)

```python
data = pd.read_csv('House_Selling_task1_part1.csv')
```

```python
data.head()
```

```python
X = data[['price', 'Size of house (sqft)']].values
X
```

```python
def initialize_centroids(X, K):
    np.random.seed(0)
    indices = np.random.choice(X.shape[0], K, replace=False)
    centroids = X[indices]
    return centroids

# Function to assign data points to the nearest centroid
def assign_clusters(X, centroids):
    distances = np.zeros((len(X), len(centroids)))
    for i in range(len(X)):
        for j in range(len(centroids)):
            distances[i, j] = np.sqrt(np.sum((X[i] - centroids[j])**2))
    return np.argmin(distances, axis=1)

# Function to update centroids based on mean of data points in each cluster
def update_centroids(X, clusters, K):
    centroids = np.zeros((K, X.shape[1]))
    for k in range(K):
        cluster_points = X[clusters == k]
        if len(cluster_points) > 0:
            centroids[k] = np.mean(cluster_points, axis=0)
    return centroids
```

```python
k = 3

centroids = initialize_centroids(X, k)
clusters = assign_clusters(X, centroids)

# Keep track of the previous centroids for convergence check
prev_centroids = np.copy(centroids)

max_iter = 100
for i in range(max_iter):
    clusters = assign_clusters(X, centroids)
    centroids = update_centroids(X, clusters, k)
    if np.array_equal(prev_centroids, centroids):
        break
    prev_centroids = np.copy(centroids)
```

```python
data['label'] = clusters
data.to_csv('House_Selling_task1_part1.csv', index=False)
```

```python
colors = ["red", "green", "blue"]
for k in range(k):
    cluster_points = X[clusters == k]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[k], s=10, label=f'Cluster {k}')

plt.xlabel("price")
plt.ylabel("Size of house (sqft)")
plt.title("price vs Size of house (sqft)")
plt.legend()
plt.show()
```

# TASK: 2 (Part 1)

```python
data = pd.read_csv('House_Selling_task1_part1.csv')

# features for clustering
X = data[['price', 'Size of house (sqft)']].values
X
```

```python
def initialize_medoids(X, K):
    np.random.seed(0)
    indices = np.random.choice(X.shape[0], K, replace=False)
    medoids = X[indices]
    return medoids
```

```python
def assign_clusters_medoids(X, medoids):
    distances = np.zeros((len(X), len(medoids)))
    for i in range(len(X)):
        for j in range(len(medoids)):
            distances[i, j] = np.abs(X[i, 0] - medoids[j, 0]) + np.abs(X[i, 1] - medoids[j, 1])  # Manhattan distance
    return np.argmin(distances, axis=1)
```

```python
def update_medoids(X, clusters, K):
    medoids = np.zeros((K, X.shape[1]))
    for k in range(K):
        cluster_points = X[clusters == k]
        if len(cluster_points) > 0:
            distances_sum = np.zeros(len(cluster_points))
            for i, point in enumerate(cluster_points):
                distances_sum[i] = np.sum(np.abs(point - cluster_points))
            medoid_index = np.argmin(distances_sum)
            medoids[k] = cluster_points[medoid_index]
    return medoids
```

```python
K = 3

medoids = initialize_medoids(X, K)
clusters = assign_clusters_medoids(X, medoids)
prev_medoids = np.copy(medoids)
```

```python
max_iter = 100
for i in range(max_iter):
    clusters = assign_clusters_medoids(X, medoids)
    medoids = update_medoids(X, clusters, K)
    if np.array_equal(prev_medoids, medoids):
        break
    prev_medoids = np.copy(medoids)
```

```python
# Plot the clusters
colors = ["red", "green", "blue"]
for k in range(K):
    cluster_points = X[clusters == k]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[k], s=10, label=f'Cluster {k}')

plt.xlabel("Price")
plt.ylabel("Size of house (sqft)")
plt.title("Price vs Size of house (sqft)")
plt.legend()
plt.show()
```
